```python
for worker_id, rows in df.groupby("WorkerId"):
    print(f"processing {worker_id}")
    output_path = os.path.join("../annotations", worker_id)
    if not os.path.exists(output_path):
        os.makedirs(output_path)
    for i, row in rows.iterrows():
        data = json.loads(row["Answer.result"])
        with open(os.path.join(output_path, f"{data['id']}.json"), "w") as f:
            json.dump(data, f, indent=4) 
```
