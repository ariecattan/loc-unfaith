```python
# save each file per topic 

for batch_result in batch_result_file:
    df = pd.read_csv(os.path.join(ANNOTATION_PATH, batch_result))
    df["sourceId"] = df["Answer.result"].apply(lambda x: json.loads(x)["sourceIds"])
    for source_id, rows in df.groupby("sourceId"):
        print(f'processing {source_id}')
        output_path = os.path.join(ANNOTATION_PATH, f'{source_id}')
        if not os.path.exists(output_path):
            os.makedirs(output_path)
        for i, row in rows.iterrows():
            input_path = f"../src/data/tofueval_chosen/mediasum/{row['Input.data']}.json"
            with open(input_path, "r") as f:
                input_data = json.load(f)
            # input_data = json.loads(html.unescape(row["Input.data"]))
            data = json.loads(row["Answer.result"])
            worker_id = row["WorkerId"]
            # print(worker_id)
            data["worker_id"] = worker_id
            for i, summary in enumerate(data["summaries"]):
                for field in ["label", "summaryId"]:
                    data["summaries"][i][field] = input_data["summaries"][i][field]
            file_name = os.path.join(output_path, f"{worker_id}.json")
            if not os.path.exists(file_name):
                with open(os.path.join(output_path, f"{worker_id}.json"), "w") as f:
                    json.dump(data, f, indent=4)
```

```python

for batch_result in batch_result_file:
    df = pd.read_csv(os.path.join(ANNOTATION_PATH, batch_result))
    for worker_id, rows in df.groupby("WorkerId"):
        print(f"processing {worker_id}")
        output_path = os.path.join(ANNOTATION_PATH, worker_id)
        if not os.path.exists(output_path):
            os.makedirs(output_path)
        for i, row in rows.iterrows():
            input_data = json.loads(html.unescape(row["Input.data"]))

            data = json.loads(row["Answer.result"])
            for i, summary in enumerate(data["summaries"]):
                for field in ["label", "summaryId"]:
                    data["summaries"][i][field] = input_data["summaries"][i][field]
            with open(os.path.join(output_path, f"{data['sourceIds']}.json"), "w") as f:
                json.dump(data, f, indent=4) 
```
