```python
# data = load_dataset("xsum", split="test")
cnn = load_dataset("cnn_dailymail", "3.0.0", split="test")
```

```python
cnn[500]
```

```python
topic = cnn[500]
source = [{"id": i, "text": token.text_with_ws} for i, token in enumerate(nlp(topic["article"]))]
sums = nlp(topic["highlights"])
summary = [{"id":i, "text": token.text_with_ws} for i, token in enumerate(sums)]
spans = [{
    "start": np.start,
    "end": np.end,
    "text": np.text
} for np in sums.noun_chunks ]
```

```python
example = {
    "source": source,
    "summary": summary,
    "spans": spans 
}
```
