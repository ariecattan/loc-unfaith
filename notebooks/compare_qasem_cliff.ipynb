```python
import json 
import os 
import pandas as pd 
import numpy as np
```

```python
# load annotatioms
cliff_path = "../annotations/prod/cliff.jsonl"
df = pd.read_json(cliff_path, lines=True)
```

```python
def get_qasem_score(qas):
    scores = [1 if sum(qa["annotations"]) <= 1 else 0 for qa in qas]
    return np.average(scores)

def get_cliff_score(labels):
    scores = [1 if label == "correct" else 0 for label in labels[0]]
    return np.average(scores)

df["qasem_score"] = df["qas"].apply(get_qasem_score)
df["cliff_score"] = df["cliff_labels"].apply(get_cliff_score)
```

```python
def get_inconsistent_qas(qas):
    return [x for x in qas if sum(x["annotations"]) >= 2]
```
